```python
# Import packages
import pandas as pd
import numpy as np
```

```python
# Load data
df = pd.read_csv('adult.csv')
df.head()
```

```python
df.info()
```

```python
# Replace '?' with NaN
df.replace('?', np.nan, inplace=True)
# Remove rows containing NaN values
df.dropna()
df.reset_index(drop=True, inplace=True)
df.isnull().sum()
```

```python
df.describe()
```

```python
df.describe(include = 'object')
```

```python
# 1. How many men and women (sex feature) are represented in this dataset?
df['gender'].value_counts()
```

```python
# 2. What is the average age (age feature) of women?
# Filter the DataFrame to include only women
women_df = df[df['gender'] == ' Female']

# Calculate the average age of women
average_age_women = women_df['age'].mean()
print(f'Average age of women is {average_age_women}')
```

```python
# 3. What is the percentage of German citizens (native-country feature)?
german_count = df[df['native-country'] ==' Germany'].shape[0]
count_total = df.shape[0]
print(f'Percentage of German citizens is {(german_count/count_total) *100}')
```

```python
# 4. What are the mean and standard deviation of age for those who earn more than 50K per year (salary feature) and those who earn less than 50K per year?

# Filter DataFrame for individuals earning more than 50K per year
high_income_df = df[df['income'] == ' >50K']

# Filter DataFrame for individuals earning less than or equal to 50K per year
low_income_df = df[df['income'] == ' <=50K']

# Calculate mean and standard deviation of age for both groups
mean_age_high_income = high_income_df['age'].mean()
std_dev_age_high_income = high_income_df['age'].std()

mean_age_low_income = low_income_df['age'].mean()
std_dev_age_low_income = low_income_df['age'].std()

print("For individuals earning more than 50K per year:")
print("Mean age:", mean_age_high_income)
print("Standard deviation of age:", std_dev_age_high_income)

print("\nFor individuals earning less than or equal to 50K per year:")
print("Mean age:", mean_age_low_income)
print("Standard deviation of age:", std_dev_age_low_income)
```

```python
# 5. Is it true that people who earn more than 50K have at least a high school
# education? (education  Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters or Doctorate feature)

# Filter DataFrame for individuals earning more than 50K per year
high_income_df = df[df['income'] == ' >50K']

# Define a list of education levels considered as at least high school
high_school_education = [' Bachelors', ' Prof-school', ' Assoc-acdm', ' Assoc-voc', ' Masters', ' Doctorate']

# Filter the high income DataFrame for individuals with education levels lower than high school
low_education_high_income_df = high_income_df[~high_income_df['education'].isin(high_school_education)]

# Check if there are any individuals in the filtered DataFrame
# If the DataFrame is empty, it means all individuals earning more than 50K have at least a high school education
if low_education_high_income_df.empty:
    print("True, all people earning more than 50K have at least a high school education.")
else:
    print("False, there are individuals earning more than 50K without at least a high school education.")
```

```python
# 6. Display age statistics for each race (race feature) and each gender (sex feature). Use groupby() and describe().
# Find the maximum age of men of Amer-Indian-Eskimo race.

# Display age statistics for each race and gender
age_stats = df.groupby(['race', 'gender'])['age'].describe()
print("Age statistics for each race and gender:")
age_stats
```

```python
# Find the maximum age of men of Amer-Indian-Eskimo race
filtered_data = df[(df['race'] == ' Amer-Indian-Eskimo') & (df['gender'] == ' Male')]
max_age_amer_indian_eskimo_men = filtered_data['age'].max()
print("Maximum age of men of Amer-Indian-Eskimo race:", max_age_amer_indian_eskimo_men)
```

```python
# 7. Among whom is the proportion of those who earn a lot (>50K) greater: married or single men (marital-status feature)?
# Consider as married those who have a marital-status starting with Married (Married-civ-spouse, Married-spouse-absent
# or Married-AF-spouse), the rest are considered bachelors.

# Filter the DataFrame for men
men_df = df[df['gender'] == ' Male']

# Categorize marital status into 'Married' and 'Single'
men_df.loc[:, 'marital_status_category'] = men_df['marital-status'].apply(lambda x: ' Married' if x.startswith(' Married') else ' Single')

# Calculate the total count of married and single men
total_married_men = men_df[men_df['marital_status_category'] == ' Married'].shape[0]
total_single_men = men_df[men_df['marital_status_category'] == ' Single'].shape[0]

# Calculate the count of married and single men earning >50K
married_men_high_earners = men_df[(men_df['marital_status_category'] == ' Married') & (men_df['income'] == '>50K')].shape[0]
single_men_high_earners = men_df[(men_df['marital_status_category'] == ' Single') & (men_df['income'] == '>50K')].shape[0]

# Calculate the proportion of high earners among married and single men
proportion_high_earners_married = married_men_high_earners / total_married_men
proportion_high_earners_single = single_men_high_earners / total_single_men

# Determine which group has a greater proportion of high earners
if proportion_high_earners_married > proportion_high_earners_single:
    print("The proportion of married men earning >50K is greater.")
elif proportion_high_earners_married < proportion_high_earners_single:
    print("The proportion of single men earning >50K is greater.")
else:
    print("The proportions of married and single men earning >50K are equal.")

men_df
```

```python
# 8. What is the maximum number of hours a person works per week (hours-per-week feature)? How many people work such a number of hours, and
# what is the percentage of those who earn a lot (>50K) among them?

# the maximum number of hours a person works per week
max_hours_per_week =df['hours-per-week'].max()

# Count the number of people who work the maximum number of hours
num_people_max_hours =  df[df['hours-per-week'] == max_hours_per_week].shape[0]

# Calculate the percentage of individuals earning >50K among those who work the maximum number of hours

percentage_of_individuals = df[(df['hours-per-week'] == max_hours_per_week) & (df['income'] == '>50K')].shape[0]

print(f'{ max_hours_per_week }\n{num_people_max_hours}\n{percentage_of_individuals}')
```

```python
# 9. Count the average time of work (hours-per-week) for those who earn a little and a lot (salary) for each country (native-country). What will these be for Japan?

avg_time = df.groupby(['income', 'native-country'])['hours-per-week'].mean()
print(avg_time,'\n\n\n')

avg_time_japan = avg_time.loc[(slice(None), ' Japan')]
avg_time_japan
```
