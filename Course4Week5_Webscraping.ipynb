# Example 1:
The Cal Poly course catalog (http://catalog.calpoly.edu/collegesandprograms/collegeofsciencemathematics/statistics/#courseinventory) contains a list of courses offered by the Statistics department. Scrape this website to obtain a DataFrame, where the unit of observation is a course. Use this DataFrame to answer the following questions:



How many 300-level courses does the Statistics department offer?

```python
#Step 1:Importing libraries
import requests
from bs4 import BeautifulSoup
```

```python
#Step 2:Adding URL
response = requests.get("https://catalog.calpoly.edu/collegesandprograms/collegeofsciencemathematics/statistics/#courseinventory")
response
```

```python
soup = BeautifulSoup(response.text,"html.parser")
soup
```

```python
#Step 3:Creating Soup object
soup = BeautifulSoup(response.content,"html.parser")
soup
```

```python
soup.find_all(class_='courseblocktitle')
```

```python
#Step 4: Retrieve the Data
soup.find(class_="courseblocktitle").text
```

```python
List1=[]

for i in soup.find_all(class_="courseblocktitle"):
  text2 = i.text[0:8]
  List1.append(text2)

List1
```

```python
import pandas as pd
df = pd.DataFrame(List1, columns=['Coursecode'])
df
```

```python
#https://pandas.pydata.org/docs/user_guide/text.html
df2 = df[df['Coursecode'].str.contains('3')]
df2.head()
```

```python
df2 = df[df['Coursecode'].str.get(5)=='3']
df2.head()
```

# **Example 2:**
Collect all the tags of the professor from this url: https://www.ratemyprofessors.com/professor/941931

```python
#Step 1:Importing libraries
import requests
from bs4 import BeautifulSoup
```

```python
#Step 2:Adding URL
response = requests.get(' https://www.ratemyprofessors.com/professor/941931')
response
```

```python
#Step 3:Creating Soup object
soup = BeautifulSoup(response.content, 'html.parser')
soup
```

```python
List_of_tags =[]
for i in soup.find_all(class_='Tag-bs9vf4-0 hHOVKF'):
  tag = i
  List_of_tags.append(tag)

List_of_tags
```

```python
Professor_tags_df= pd.DataFrame(List_of_tags, columns =['Tags'])
Professor_tags_df
```

```python
# 1.	Write a Python program to test if a given page is found or not on the server.

import requests

def test(url):
  response = requests.get(url)
  if response.status_code == 200:
    return f'Th page({url}) found'

  else:
    return f'Th page({url}) not found'

test('https://www.ratemyprofessors.com/professor/941931')
```

```python
# 2.	 Write a Python program to extract and display all the header tags from en.wikipedia.org/wiki/Main_Page

import requests
from bs4 import BeautifulSoup

url ='https://en.wikipedia.org/wiki/Main_Page'

response = requests.get(url)

soup = BeautifulSoup(response.content, 'html.parser')

tags =['h1', 'h2', 'h3']
header_tag = soup.find_all(tags)

for i in header_tag:
  print(i.text)
```
