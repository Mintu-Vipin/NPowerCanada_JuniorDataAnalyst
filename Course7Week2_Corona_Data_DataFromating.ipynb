```python
# importing necessary packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load the excel data to the dataframe
df=pd.read_excel('Corona_data.xlsx')
df
```

```python
# Print no of rows and columns
df.shape
```

```python
# dataframe information
df.info()
```

```python
# print summary
df.describe()
```

```python
# print the null values
df.isnull().sum()
```

```python
# drop the column 'date'
df.dropna(subset=['date'], axis=0, inplace=True)
```

```python
# cases - replace the value of NaN with median(mean>median)
df['cases']=df['cases'].fillna(df['cases'].median())
df
```

```python
# active cases - replace the value of NaN with median(mean>median)
df['active cases']=df['active cases'].fillna(df['active cases'].median())
df
```

```python
# find the rows with 'deaths'=-1 and drop it

df =df[df['deaths']!= -1]
df
```

```python
df.describe()
```

```python
# find the rows with 'deaths'= NaN and replace it with median

df['deaths']=df['deaths'].fillna(df['deaths'].median())
df
```

```python
df.isnull().sum()
```

```python
# cases - Creating the bins
cases_bins = np.linspace(min(df['cases']), max(df['cases']), 10)

# deaths - Creating the bins
active_cases_bins = np.linspace(min(df['active cases']), max(df['active cases']), 10)

# deaths - Creating the bins
deaths_bins = np.linspace(min(df['deaths']), max(df['deaths']), 10)
```

```python
# Create Histograms
plt.figure(figsize=(10, 6))

# Histogram for 'cases'
plt.subplot(3, 1, 1)
plt.hist(df['cases'], bins=cases_bins, edgecolor='black')
plt.xlabel('Cases')
plt.ylabel('Frequency')
plt.title('Histogram of Cases with Bins')

# Histogram for 'active cases'
plt.subplot(3, 1, 2)
plt.hist(df['active cases'], bins=active_cases_bins, edgecolor='black')
plt.xlabel('Active Cases')
plt.ylabel('Frequency')
plt.title('Histogram of Active Cases with Bins')

# Histogram for 'deaths'
plt.subplot(3, 1, 3)
plt.hist(df['deaths'], bins=deaths_bins, edgecolor='black')
plt.xlabel('Deaths')
plt.ylabel('Frequency')
plt.title('Histogram of Deaths with Bins')

plt.tight_layout()
plt.show()
```
